# **MACCSkeys**

```python
from rdkit import Chem
from tqdm import tqdm
import numpy as np
```

```python
from rdkit.Chem import MACCSkeys
```

```python
def generate_MACCS(data):
    MACCS_keys = []
    mols = [Chem.MolFromSmiles(x) for x in data if x is not None]
    for mol in tqdm(mols):
        maccs = list(MACCSkeys.GenMACCSKeys(mol).ToBitString())
        MACCS_keys.append(maccs)
    return np.array(MACCS_keys)
```

```python
MACCS_keys = generate_MACCS(df['SMILES_uncharge'])
```

```python
for i, row in df.iterrows():
    for j in range(MACCS_keys.shape[1]):
        df.at[i, f'MAC_{j+1}'] =  MACCS_keys[i][j]

df.head()
```

# **Avalon**

```python
from rdkit.Avalon import pyAvalonTools
```

```python
def generate_AVfpts(data):
    Avalon_fpts = []
    mols = [Chem.MolFromSmiles(x) for x in data if x is not None]
    for mol in tqdm(mols):
        avfpts = pyAvalonTools.GetAvalonFP(mol, nBits=512)
        Avalon_fpts.append(avfpts)
    return np.array(Avalon_fpts)
```

```python
Avalon_fpts = generate_AVfpts(df['SMILES_uncharge'])
```

```python
Avalon_fpts.shape
```

```python
for i, row in df.iterrows():
    for j in range(Avalon_fpts.shape[1]):
        df.at[i, f'A_{j+1}'] = Avalon_fpts[i][j]

df.head()
```

```python
import sklearn
from sklearn.feature_selection import SelectKBest, mutual_info_regression
from sklearn.model_selection import KFold, GridSearchCV, cross_validate
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
del df['SMILES_uncharge']
```

# **dvfd,**

```python
def select_features(X, y):
    fs = SelectKBest(score_func=mutual_info_regression, k=133)
    fs.fit(X, y)
    X_fs = fs.transform(X)
    return X_fs, fs
```

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
X_fs, fs = select_features(X, y)
```

```python
cols_idxs = fs.get_support(indices=True)
df = df.iloc[:, cols_idxs]
```

```python
df.head()
```

```python
split = KFold(n_splits=5, random_state=41, shuffle=True)
scores= {
         "Q2": "r2",
         "MSE": make_scorer(mean_squared_error, squared=False)
         }
```

```python
X_train, X_test, y_train, y_test = X, X, y, y
```

```python
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
MLR_model = XGBRegressor()
```

# **Metrics**

```python
MLR_model = XGBRegressor(random_state=42, n_estimators=1500, max_depth=10, learning_rate=0.01, subsample=0.5, tree_method='gpu_hist', gpu_id=0)
```

```python
cv_scores = cross_validate(MLR_model, X_train, y_train, scoring=scores, cv=split)
print(f"On cross-validation:")
print(f"Mean RMSE score is {cv_scores['test_MSE'].mean().round(3)} ± {cv_scores['test_MSE'].std().round(3)}")
print(f"Mean Q2 score is {cv_scores['test_Q2'].mean().round(3)} ± {cv_scores['test_Q2'].std().round(3)}")
```

```python
MLR_model.fit(X_train, y_train)
y_pred = MLR_model.predict(X_test)

print(f"R2: {r2_score(y_test, y_pred).round(3)}")
```
